```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("10-k-filingdata.csv")
```

```python
# Check unique values first
print(df['Fiscal Year'].unique())

# Convert to integer (if appropriate)
df['Fiscal Year'] = df['Fiscal Year'].astype(str)
```

```python
df.head()
```

```python
numeric_cols = ['Total Revenue', 'Net Income', 'Total Assets', 'Total Liabilities', 'CFO Activities']
df[numeric_cols] = df[numeric_cols].replace(',', '', regex=True).astype(float)
```

```python
df = df.sort_values(by=['Company', 'Fiscal Year'], ascending=[True, False])
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
```

```python
df.head()
```

```python
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

```python
df.head()
```

```python
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change(-1) * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change(-1) * 100
df['CFO Growth (%)'] = df.groupby('Company')['CFO Activities'].pct_change(-1) * 100
```

```python
df.head()
```

```python
sns.lineplot(data=df, x='Fiscal Year', y='Total Revenue', hue='Company')
plt.title("Revenue Trend by Company")
plt.gca().invert_xaxis()  # Show years from past to present
plt.show()
```
